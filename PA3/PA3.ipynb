```python
import numpy as np
import random
import math

train_file = open("pa3train.txt","r")
train = [line.strip() for line in train_file]
train = [[int (i) for i in line.split()] for line in train]


test_file = open("pa3test.txt","r")
test = [line.strip() for line in test_file]
test = [[int (i) for i in line.split()] for line in test]

train_label = []
# 1 for label 1, -1 for label 2 
for i in range(len(train)):
    if(train[i][-1] == 1):
        train_label += [1]
    else:
        train_label += [-1]

test_label = []
# 1 for label 1, -1 for label 2 
for i in range(len(test)):
    if(test[i][-1] == 1):
        test_label += [1]
    else:
        test_label += [-1]

train = [line[:len(line)-1] for line in train]
test = [line[:len(line)-1] for line in test]

```

```python
def perceptronKerPoly(data,label,M):
    for t in range(len(data)):
        s = 0
        for i in range(len(M)):
            index = M[i]
            s += label[index]*((np.dot(data[index],data[t])+10)**2)
        s = label[t]*s
        if(s <= 0):
            M += [t]
```

```python
# check error for KerPoly
M = []
k = 0
train_error_rate = []
test_error_rate = []
while(k < 5):
    perceptronKerPoly(train,train_label,M)
    r = 0
    train_error_num = 0
    for i in range(len(train)):
        r = 0
        for j in range(len(M)):
            r += train_label[M[j]]*((np.dot(train[M[j]],train[i])+10)**2)
        if(r > 0): r = 1
        elif(r < 0): r = -1
        else: r = [random.choice([-1,1])]

        if(r != train_label[i]): 
            train_error_num += 1
    train_error_rate += [train_error_num/len(train)]

    r = 0
    test_error_num = 0
    for i in range(len(test)):
        r = 0
        for j in range(len(M)):
            r += train_label[M[j]]*((np.dot(train[M[j]],test[i])+10)**2)

        if(r > 0): r = 1
        elif(r < 0): r = -1 
        else: r= [random.choice([-1,1])]    

        if(r != test_label[i]): 
            test_error_num += 1
    test_error_rate += [test_error_num/len(test)]
    k += 1

print("For the first pass of poly kernel function")
print("train error rate is: ",train_error_rate[0])
print("test error rate is: ",test_error_rate[0])
print("For the second pass")
print("train error rate is: ",train_error_rate[1])
print("test error rate is: ",test_error_rate[1])
print("For the third pass")
print("train error rate is: ",train_error_rate[2])
print("test error rate is: ",test_error_rate[2])
print("For the fourth pass")
print("train error rate is: ",train_error_rate[3])
print("test error rate is: ",test_error_rate[3])
print("For the fifth pass")
print("train error rate is: ",train_error_rate[4])
print("test error rate is: ",test_error_rate[4])
```

```python
def perceptronKerExpo(data,label,M):
    for t in range(len(data)):
        s = 0
        for i in range(len(M)):
            index = M[i]
            #ker = np.subtract(data[index],data[t]) 
            ker = -1*(np.linalg.norm(np.array(data[index])-np.array(data[t])))/20 
            ker = math.exp(ker)
            s += label[index]*(ker)
        s = label[t]*s
        if(s <= 0):
            M += [t]
```
